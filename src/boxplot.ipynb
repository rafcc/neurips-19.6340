```python
import numpy as np
import subfunction
import glob
import yaml
import matplotlib.pyplot as plt
```

```python
def summarize_results(target_dir):
    out_dic = {}
    criteria = ['random_l2risk']
    methods = ['borges','inductive']
    for c in criteria:
        out_dic[c] = {}
        for m in methods:
            out_dic[c][m] = []
    for d in glob.glob(target_dir+"*/*"):
        fname = d +'/output.yml'
        yml = yaml.load(open(fname,'r'))
        for c in criteria:
            for m in methods:
                out_dic[c][m].append(yml['results'][c][m])
    return(out_dic)
summarize_results(target_dir="../results/linear_M.3_L.3_D.2_N.1000_sigma.0.1")
```
